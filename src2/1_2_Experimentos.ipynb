```python
c = 0

for maze in mazes:
    all_transitions = transition_generator(maze["layout"])

    for func in transition_funcs:

        for i in n_models:
            N = i
            if func == "REPETITION":
                new_transitions = transitions_families_generator_WITH_REPETITION(all_transitions, N)
            elif func == "FULLY":
                new_transitions = transitions_families_generator_FULLY_REPRESENTED(all_transitions, N)

            for epochs in n_epochs:

                for test_split in test_splits:
                    c += 1
                    for j in range(N):
                        path = "../data/ensembles/" + maze["name"] + "/" + func + "/" + str(N) + "_models/" + str(epochs) + "_epochs/" + str(test_split) + "_test_split"
                        #Need to create the path folder if it does not exist
                        if not os.path.exists(path):
                            os.makedirs(path)

                        model = ModelTrainer(j, shape, new_transitions[j], epochs)
                        model.train(test_size=test_split, path=path)
```
